```python
#importing packages
import requests
from bs4 import BeautifulSoup
import time
!pip install Ipython
from IPython.display import clear_output
```

```python
# bring things together

#create your empty lists
House_description = []
Address_of_property = []
Agent_of_property = []
Add_update_date_of_property = []
Price_of_property = []
Number_of_rooms = []
Num_of_bathrooms = []
Num_of_toilets = []


#intialising
base_url = "https://www.privateproperty.com.ng/houses-for-sale/enugu?page="
for i in range(0,9):
  url = (base_url+str(i))

  #we are pausing for 3 seconds (to prevent jamming the website)
  time.sleep(3)

  #showing the page being scaped and clearing the pages
  print(f'scrapping info from page {i}')
  clear_output(wait = True)

  #getting url and storing it.
  response = requests.get(url)
  response = response.content
  soup = BeautifulSoup(response, 'html.parser')

  #get the container of all the houses
  container = soup.find('div', class_ = 'result-listings')

  #enter into container and find all the house info and store them
  house_info = container.find_all('div', class_ = 'similar-listings-info')
  house_info2 = container.find_all('div', class_ = 'similar-listings-price')

  #get specific house description from the house_info
  for house in house_info:
    house_information = house.find('h2')
    House_description.append(house_information.text)
  
  #get specific address of property
  for address in house_info:
    house_address = address.find("p", class_ = "listings-location")
    Address_of_property.append(house_address.get_text("n/", strip = True))
  
  #get specific agent of property
  for agent in house_info:
    house_agent = agent.find("img")
    Agent_of_property.append(house_agent.attrs["alt"])

  #get specific dates for addding and updating of property online
  for date in house_info:
    house_date = date.find('h5', class_="mt-0")
    Add_update_date_of_property.append(house_date.text)
  
  #get specific prices of property
  for price in house_info2:
    house_price = price.h4
    house_price = house_price.contents[3]
    Price_of_property.append(house_price.text)
  
  # get number of rooms per property
  for room in house_info2:
    num_rooms = room.find("ul", class_="property-benefit")
    num_rooms = num_rooms.li
    Number_of_rooms.append(num_rooms.get_text("n/", strip = True))

  # get number of bathrooms per property
  for item in house_info2:
    bathroom = item.contents[3]
    bathroom = bathroom.contents[3]
    Num_of_bathrooms.append(bathroom.get_text("n/", strip = True))

  # get number of toilets per property
  for item in house_info2:
    toilet = item.contents[3]
    toilet = toilet.contents[5]
    Num_of_toilets.append(toilet.get_text("n/", strip = True))










```

```python
import pandas as pd
# Calling DataFrame constructor after zipping
# both lists, with columns specified
house= ['Property_description', 'Address_of_property','Agent_of_property','Add_update_date_of_property',
                           'Price_of_property', 'Number_of_rooms', 'Num_of_bathrooms','Num_of_toilets']
df = pd.DataFrame(list(zip(House_description, Address_of_property,Agent_of_property,Add_update_date_of_property,
                           Price_of_property, Number_of_rooms, Num_of_bathrooms,Num_of_toilets)),
			columns = house)
df
```
